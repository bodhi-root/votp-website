# Yoga Sutras File Prep

This notebook prepares a JSON file that lets us show verses from the Yoga Sutras. I found a [spreadsheet with the Yoga Sutras](https://docs.google.com/spreadsheets/d/1HrfWM7KeBPHHOCXbpKMVk3oTV9nrOZYr/edit?usp=drive_link&ouid=101257354704441160839&rtpof=true&sd=true) from a Sanskrit student on [this Reddit post](https://www.reddit.com/r/yoga/comments/17zf5co/yoga_sutras_in_excel_with_wordforword_translation/). This was a great starting point!

All I had to do to get started was read it in, rename some columns, and write it back out as JSON. Maybe I'll add to it later. It might be good to add more translations that I like.

NOTE: I did not run this in the project venv. I used a separate one that had `polars` and `fastxml` installed.

````python
import polars as pl
````

````python
#df = pl.read_excel("YogaSutraDecomposed.xlsx", sheet_name="Ch1")
#df.head()
````

````python
data = []
for sheet in ["Ch1", "Ch2", "Ch3", "Ch4"]:
    df = pl.read_excel("YogaSutraDecomposed.xlsx", sheet_name=sheet)
    display(df.head())
    data.append(df)    
````

I don't know why the spreadsheet calls it "Brahmi". Brahmi is an ancient writing system that is definitely not this one. This one is clearly devanagari. IAST is the "International Alphabet of Sanskrit Transliteration." Note that "Sanskrit" refers to the spoken language. Devanagari, IAST, and even Brahmi are different writing systems for the same spoken words and sounds.

````python
def clean(df):
    df = df.rename({
        "Book": "pada",
        "Sutra": "sutra",
        "Brahmi": "devanagari",
        "IAST": "iast",
        "English": "english"
    })
    return df.select(["pada", "sutra", "devanagari", "iast", "english"])

cleaned = [clean(df) for df in data]

all_df = pl.concat(cleaned, how="vertical")
all_df = all_df.with_columns(index=pl.arange(1, all_df.height + 1))
````

````python
import json

all_df = all_df.select(["index", "pada", "sutra", "devanagari", "iast", "english"])
json_obj = all_df.to_dicts()

with open("yoga_sutras.json", "w") as fp:
    json.dump(json_obj, fp, ensure_ascii=False, indent=2)

#json_obj
````
